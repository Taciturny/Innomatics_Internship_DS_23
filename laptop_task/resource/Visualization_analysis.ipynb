```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
from plotly.subplots import make_subplots
import plotly.express as px
```

```python
vis = pd.read_csv('df.csv')
vis.head()
```

```python
sns.boxplot(y=vis["price"]);
```

```python
columns = ["price", "ram_size", "ram_type", "storage", "storage_type", "os_type"]

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

for col, ax in zip(columns, axes.flatten()):
    sns.violinplot(data=vis, x=col, ax=ax, inner="box", palette="Set3", cut=2, linewidth=3)
    ax.set_title(col)
    ax.set_xlabel('')
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
import pandas as pd

# Define the data
columns = ["price", "ram_size", "ram_type", "storage", "storage_type", "os_type"]

# Define the traces
traces = []
for column in columns:
    trace = go.Scatter(x=vis[column], y=vis['price'], mode='markers', name=column)
    traces.append(trace)

# Define the layout
layout = go.Layout(title='Price vs. Features', xaxis_title='Features', yaxis_title='Price')

# Create the figure
fig = go.Figure(data=traces, layout=layout)

# Display the figure
pyo.plot(fig, filename='price_vs_features.html')
```

```python
def multivariate_plot():
    fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5), width_ratios=[1, 1, 1, 1.5], gridspec_kw={'wspace': 0.2, 'hspace': 0.4})
    
    # Create box plots for Ram Size, Storage, StorageType, and OS
    sns.boxplot(data=vis, x='ram_size', y='price', ax=axes[0])
    sns.boxplot(data=vis, x='storage', y='price', ax=axes[1])
    sns.boxplot(data=vis, x='storage_type', y='price', ax=axes[2])
    sns.boxplot(data=vis, x='os_type', y='price', ax=axes[3])

    # Set titles for each plot
    axes[0].set_title("Ram Size vs Price", fontsize=14)
    axes[1].set_title("Storage vs Price", fontsize=14)
    axes[2].set_title("Storage Type vs Price", fontsize=14)
    axes[3].set_title("OS vs Price", fontsize=14)

    # Remove x-labels for all subplots except the last one
    for i in range(len(axes)):
        if i != len(axes) - 1:
            axes[i].set_xlabel('')
    
    # Set y-label for the first plot only
    axes[0].set_ylabel("Price", fontsize=14)
    
    # Set font size for all tick labels
    for ax in axes:
        ax.tick_params(labelsize=12)
    
    # Set overall title for the figure
    fig.suptitle("Multivariate Plots", fontsize=18, fontweight='bold', y=1.1)
    
    # Display the figure in Jupyter notebook
    plt.show()

```

```python
multivariate_plot()
```

```python
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import streamlit as st

def multivariate_plot():
    fig = make_subplots(rows=1, cols=4, subplot_titles=("Ram Size vs Price", "Storage vs Price", "StorageType vs Price", "OS vs Price"), 
                        horizontal_spacing=0.1, vertical_spacing=0.1)

    # Create box plots for Ram Size, Storage, StorageType, and OS_type
    fig.add_trace(px.box(vis, x='ram_size', y='price').data[0], row=1, col=1)
    fig.add_trace(px.box(vis, x='storage', y='price').data[0], row=1, col=2)
    fig.add_trace(px.box(vis, x='storage_type', y='price').data[0], row=1, col=3)
    fig.add_trace(px.box(vis, x='os_type', y='price').data[0], row=1, col=4)

    # Update the layout
    fig.update_layout(height=600, width=800, showlegend=False, font_color="blue", title="Multivariate plots", 
                      title_font_size=30, title_font_color="red", font=dict(size=18, family='Arial'))

    return fig

# Display the figure in Streamlit
st.plotly_chart(multivariate_plot())
```
